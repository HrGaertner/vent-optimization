```python
def relative_to_absolute(humidity, temperature):
    # Source https://journals.ametsoc.org/view/journals/bams/86/2/bams-86-2-225.xml?tab_body=pdf Equation 6 p.226
    C_1 = 610.94 #Kp
    A_1 =  17.625
    B_1 =243.04 #°C
    return ((humidity/100)*C_1*np.exp((A_1*temperature)/(B_1+temperature)))/1000

absolute = []
for index,row in shower.iterrows():
    absolute.append(relative_to_absolute(float(row["humidity"]), float(row["temperature"])))

shower["absolute"] = absolute
```

```python
fix, ax = plt.subplots()
shower.plot.scatter(x="local time", y="absolute", xlabel="Zeit in Minuten", ylabel="absolute Luftfeuchtigkeit", ax=ax, c="Id", colormap="viridis_r")
```

```python
def e_s(T): # Source https://journals.ametsoc.org/view/journals/bams/86/2/bams-86-2-225.xml?tab_body=pdf Equation 6 p.226
    C_1 = 610.94 #Kp
    A_1 =  17.625
    B_1 =243.04 #°C
    return C_1*np.exp((A_1*T)/(B_1+T))

def h(h_0, h_out, F_größe, R_größe, t, k_1, k_2):  # The absolute humidity model
    return (h_0 -h_out)/((t+1)**((k_1*F_größe)/(k_2*R_größe))) + h_out

def T(T_0, T_out, F_größe, R_größe, t, k_1, k_2): # The temperature model
    return (T_0-T_out)/((t+1)**((k_1*F_größe)/(k_2*R_größe))) + T_out

def model_h(window, room, L_out, L_begin, T_out, T_begin,t, k1,k2,k3, k4, training=False):
    L_abs_begin = (L_begin/100)*e_s(T_begin)
    L_abs_out = (L_out/100)*e_s(T_out)
    pred_T = T(T_begin, T_out, window, room, t, k3, k4)
    pred_L = h(L_abs_begin, L_abs_out, window, room, t, k1, k2)
    result= (pred_L/e_s(pred_T))*100
    
    if training:
        to_return = []
        for res in result:
            if res >100:
                to_return.append(-np.inf)
            else:
                to_return.append(res)
        return to_return
    return result

constants = [49.9737675 ,  2.28452262, 49.3679433 ,  8.39747826]
```

```python
y = []
x = []
for i in range(200):
    y.append(max(vent[vent["local time"] > i]["humidity_bathroom_middle"]))
    x.append(i)

# setting the axes at the centre
# plot the function
plt.plot(x,y, 'r')
plt.xlabel("Zeit")
plt.ylabel("Maximale Luftfeuchtigkeit über alle Messdaten")

# show the plot
plt.show()
```

```python
Id = 7

f_val =  vent[(vent["Id"]==Id) & (vent["local time"] == 0)]
code_mid = {1: 0.8*0.95, 2: 2*0.8*0.95, 3: 0.95*0.15 + 0.80*0.005, 4: 2*(0.95*0.15 + 0.80*0.005)}

ax = vent[vent["Id"] == Id].plot.scatter(x="local time", y="humidity_outside")
vent[vent["Id"] == Id].plot.scatter(x="local time", y="humidity_bathroom_middle", ax=ax, color="red")

print(code_mid[f_val["Config"].values[0]], 45, f_val["humidity_outside"].values[0], f_val["humidity_bathroom_middle"].values[0], f_val["temperature_outside"].values[0], f_val["temperature_bathroom_middle"].values[0])

x = np.linspace(0,140,500)
y =  model_h(code_mid[f_val["Config"].values[0]], 45, f_val["humidity_outside"].values[0], f_val["humidity_bathroom_middle"].values[0], f_val["temperature_outside"].values[0], f_val["temperature_bathroom_middle"].values[0],x, *constants)

plt.plot(x,y, 'y')
plt.show()
```
